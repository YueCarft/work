```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\YUE\Desktop\interface2.csv")
```

```python
data=data["Rating"]
```

```python
count_1=0
count_2=0
count_3=0
count_4=0
count_5=0
for i in data:
    if i==1:
        count_1+=1
    if i==2:
        count_2+=1
    if i==3:
        count_3+=1
    if i==4:
        count_4+=1
    if i==5:
        count_5+=1
#print(count_1)
#print(count_2)
#print(count_3)
#print(count_4)
#print(count_5)
print(count_1+count_2+count_3)
print(count_4+count_5)
all_reviews=count_1+count_2+count_3+count_4+count_5
#print(all_reviews)
```

```python
x=['num_1','num_2',"num_3","num_4","num_5"]
y=[count_1,count_2,count_3,count_4,count_5]
plt.figure(figsize=(20,10),dpi=80)
plt.xticks(range(len(x)),x)
plt.bar(range(len(x)),y,width=0.3)
index=np.arange(len(x))
for a,b in zip(index,y):   
    plt.text(a,b,'%.0f'%b,ha='center',va='bottom',fontsize=12);
plt.title("The number of stars from one to five")
plt.show()
```
